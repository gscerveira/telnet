```python
# Cell 1 - Install SSH server and cloudflared
!apt-get update -qq
!apt-get install -qq openssh-server > /dev/null

# Configure SSH to allow root login
!echo "PermitRootLogin yes" >> /etc/ssh/sshd_config
!echo "PasswordAuthentication yes" >> /etc/ssh/sshd_config

# Install cloudflared
!wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64.deb
!dpkg -i cloudflared-linux-amd64.deb > /dev/null

print("SSH and cloudflared installed!")
```

```python
# Cell 2 - Set password and start SSH
import subprocess
import random
import string

# Generate random password
password = ''.join(random.choices(string.ascii_letters + string.digits, k=12))

# Set root password
subprocess.run(['chpasswd'], input=f'root:{password}'.encode())

# Start SSH service
!service ssh start

print(f"Root password: {password}")
```

```python
# Cell 3 - Start cloudflared tunnel
# This cell will keep running - copy the URL that appears
!cloudflared tunnel --url ssh://localhost:22 2>&1 | grep -Eo 'https://[a-zA-Z0-9-]+\.trycloudflare\.com'
```
